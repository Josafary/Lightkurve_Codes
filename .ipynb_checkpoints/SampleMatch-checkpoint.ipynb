```python
import pandas as pd
import numpy as np
```

```python
table1 = pd.read_csv('DR3data/383TicList.txt')
table2 = pd.read_csv('Schaffenroth2023_Sample.csv', sep=';')
#table2 = pd.read_csv('386_OB_Stars/LCs_estrelasOB/get_tspan_merged.csv', sep=';')

l1 = []
l2 = []
for i in range(len(table1)):
    l1.append(table1['TIC'][i])
    
for j in range(len(table2)):
    l2.append(table2['TIC'][j])
    

finalList = []
for i in range(len(l1)):
    for j in range(len(l2)):
        if l1[i] == l2[j]:
            finalList.append(l2[j])
            
            
colunas = ['TIC', 'Match']
df = pd.DataFrame(columns=colunas)
for i in range(len(l1)):
    df = df.append({'TIC':int(l1[i]), 'Match':'NO'}, ignore_index=True)

for i in range(len(l1)):
    for j in range(len(finalList)):
        if df['TIC'][i] == finalList[j]:
            df['Match'].iloc[i] = 'YES'
        else:
            pass
        
df.to_csv('SampleMatch_Schaffenroth.csv', sep='|', index=False)
print('DONE!')
```
