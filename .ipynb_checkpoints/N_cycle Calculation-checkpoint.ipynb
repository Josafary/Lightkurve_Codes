```python
import pandas as pd
```

```python
file = pd.read_csv('../../NAOS/Bstars/DR3data/Tables/New_table_rot.csv', sep=';')
file
```

```python
N_cyc = []

for i in range(len(file)):
    n_cycle = file.t_span[i] / file.Period[i]
    N_cyc.append(round(n_cycle, 3))

N_cyc
```

```python
N_cycle = pd.DataFrame(N_cyc, columns=['N_cycle'])
New_table = pd.concat([file, N_cycle], axis=1)
New_table
```
