```python
import os
import pandas as pd
from pptx import Presentation
from pptx.util import Inches

# Create a DataFrame from the sdB table
df = pd.read_excel('/../../NAOS/Bstars/DR3data/sdb_new_table.xlsx')

# Create a list with all files in the given directory
path = '/../../NAOS/Bstars/DR3data/MAPS_Files/dat_sdb_all/png_coroect/'
file_list = os.listdir(path)

file_listP = []
file_listU = []

for i in range(0, len(file_list)):
    file_listP.append(file_list[i].split('.')[0])
    
for j in range(0, len(file_list)):
    file_listU.append(file_listP[j].split('_')[0])
    
# Create ppt presentation to add slides to
ppt = Presentation()

# FIRST FRAME
frame1_register = ppt.slide_layouts[0]
frame1 = ppt.slides.add_slide(frame1_register)
title1 = frame1.shapes.title
title1.text = 'MY PRESENTATION'
author = frame1.placeholders[1]
author.text = 'by Josafary Campelo'

# Create one frame for each image in the directory, with it ID_Number as title for each frame
for k in range(0, len(file_list)):
    frame_register = ppt.slide_layouts[5]
    frame = ppt.slides.add_slide(frame_register)
    
    title = frame.shapes.title
    title.text = file_listP[k]
    
    img = path + str(file_list[k])
    
    img_left = Inches(0.5)
    img_top = Inches(1.6)
    img_width = Inches(9)
    img_height = Inches(4.5)
    
    add_picture = frame.shapes.add_picture(img, img_left, img_top, img_width, img_height)
    
    left = Inches(0.5)
    top = Inches(6.3)
    width = Inches(9)
    height = Inches(1)
    for l in range(0, len(df)):
        if str(df.TIC[l]) == file_listU[k]:
            params = frame.shapes.add_textbox(left, top, width, height)
            params.text = 'Our Class: '+str(df.Our_Class[l]) + '  |  ' + 'Period: '+str(df.Period_d[l]) + '  |  ' + 'ePeriod: '+str(df.ePeriod[l]) + '  |  ' + 't_span: '+str(df.t_span[l]) + '\n' + 'Teff: '+str(df['Teff (K)'][l]) +  '  |  ' + 'log(g): '+str(df['log(g) (cm/s^2)'][l]) + '  |  ' + 'Mass: '+str(df['Mass (Msun)'][l]) + '\n' + 'Simbad Class: '+str(df.Simbad_Class[l]) + '  |  ' + 'Spectral Type: '+str(df['Spectral Type'][l])
        
    
    
    
ppt.save('presentation_MAPS.pptx')
```
